```python
import os
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.prompts import ChatPromptTemplate
from google.colab import userdata
```

```python
os.environ["GOOGLE_API_KEY"] = userdata.get('GEMINI_API_KEY')

llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash",
    temperature=0)
```

```python
prompt_template = ChatPromptTemplate.from_messages(
    [
        ("system", """You are a prompt expert engineer. Your task is to evaluate the quality of a user prompt.

        ###Prompt Quality Criterion
        Score the prompt using the following prompt quality criteria (0-10 each):
        1. Clarity - Checks whether the prompt is easy to understand and has a clear goal.
        2. Specificity/Details - Evaluates whether sufficient details and requirements are provided.
        3. Context - Checks if background information, audience, or use case is mentioned.
        4. Output Format & Constraints - Checks whether expected output format, tone, or length is specified.
        5. Persona Defined - Confirms whether a prompt assigns a specific role.

        Then compute the Final Score = average of all five.

        ###Output:
        - Final Score (0-10)
        - Scores of each quality criterion
        - Short Explanation
        - 2-3 suggestions to improve the prompt
        """),
        ("user", "Enter the prompt: {inputPrompt}")
    ]
)
```

```python

prompt_value = prompt_template.format_prompt(inputPrompt="You are an expert developer and reviewer. Analyze my code, fix all issues, improve performance, and rewrite it if needed. Make it clean, scalable, and perfect. Add comments and explanations but don’t make it too long. Ensure it handles all edge cases and works everywhere.")
```

```python
response=llm.invoke(prompt_value)
print(response.content)
```
